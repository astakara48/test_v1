```python
df1 = pd.read_csv('머드 로션.csv', encoding='utf-8', engine='python')
df2 = pd.read_csv('머드 마스크팩.csv', encoding='utf-8', engine='python')
df3 = pd.read_csv('머드 바드클렌저.csv', encoding='utf-8', engine='python')
df4 = pd.read_csv('머드 컨디셔너.csv', encoding='utf-8', engine='python')
df5 = pd.read_csv('머드 크렌징.csv', encoding='utf-8', engine='python')
df6 = pd.read_csv('머드 크림.csv', encoding='utf-8', engine='python')
df7 = pd.read_csv('머드 파우더.csv', encoding='utf-8', engine='python')
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.drop_duplicates(['blogurl']).reset_index(drop=True)
```

```python
df
```

```python
df = df[df['title'].notnull()]
df = df[df['blogurl'].notnull()]
df = df[df['writedate'].notnull()]
df = df[df['maincontant'].notnull()]
df = df.reset_index(drop=True)
```

```python
df['writedate'] = df['writedate'].astype(int)
```

```python
def get_date(x):
    x = str(x)
    date = ''
    if len(x) == 7:
        year = str(x[:4])
        month = str(0) + str(x[4])
        day = str(x[-2:])
        date = year + month + day
    else:
        date = x
        
    return date
```

```python
test_df = df.copy()
```

```python
test_df['writedate'] = test_df['writedate'].apply(lambda x: get_date(x))
```

```python
test_df['writedate'].unique()
```

```python
test_df = test_df.sort_values(['writedate'], ascending=False).reset_index(drop=True)
```

```python
test_df
```

```python
test = '● 데일리멜로 ● 포브리에 선크림 / 아이도쓰는 순한선크림 핫딜 (업그레이드 되었어..'

re.sub('[^0-9-a-zA-Z가-힣 .]+', '', test)
```

```python
def normalize(x):
    return re.sub('[^0-9-a-zA-Z가-힣 .]+', '', x)
```

```python
test_df['title'] = test_df['title'].apply(lambda x: normalize(x))
```

```python
test_df['maincontant'] = test_df['maincontant'].apply(lambda x: normalize(x))
```
